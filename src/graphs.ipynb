```python
import pandas as pd
import matplotlib.pyplot as plt   
```

```python
results150 = pd.read_csv('results150.csv')
results150.head()
```

```python
results300 = pd.read_csv('results300.csv')
results300.head()
```

```python
results450 = pd.read_csv('results450.csv')
results450.head()
```

```python
results600 = pd.read_csv('results600.csv')
results600.head()
```

```python
plt.plot(results150["Frame"], results150["fifo_faults"], label="FIFO")
plt.plot(results150["Frame"], results150["aging_faults"], label="Envelhecimento")
plt.xlabel("Moldura")
plt.ylabel("Falta de pgina")
plt.xticks(range(0, results150["Frame"].max() + 1, 25))
plt.title("Faltas de pgina entre o FIFO e o Envelhecimento com 150 pginas")
plt.legend()
plt.show()
```
